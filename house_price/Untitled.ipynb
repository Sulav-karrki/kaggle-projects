```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
    
```

```python
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
cat_pipeline = Pipeline([
    ("encoder", OneHotEncoder(handle_unknown="ignore")),
    ("cat_imputer", SimpleImputer(strategy="median"))
])

        
```

```python
num_attr = list(X_train.select_dtypes(include = "number"))
cat_attr = list(X_train.select_dtypes(include = "object"))
print(len(num_attr))
print(len(cat_attr))
print(X_train.shape)
```

```python
from sklearn.compose import ColumnTransformer
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attr),
    ("cat", cat_pipeline, cat_attr)
])
X_train_processed = full_pipeline.fit_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train_processed, y_label)
```

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error
prediction = lr.predict(X_train_processed)
print("MSE:", mean_squared_error(y_label, prediction))
print("RMSE:", root_mean_squared_error(y_label, prediction))
```

```python
test = pd.read_csv("test.csv")
test_l = pd.read_csv("sample_submission.csv")
test_label = test_l["SalePrice"]
X_test_processed = full_pipeline.transform(test)
```

```python
y_pred = lr.predict(X_test_processed)
```

```python
print("MSE:", mean_squared_error(test_label, y_pred))
print("RMSE:", root_mean_squared_error(test_label, y_pred))
```

```python
submission = pd.DataFrame({
    "Id" : test["Id"],
    "SalePrice" : y_pred
})
```

```python
submission.to_csv("sample_submission.csv", index =False)
```
